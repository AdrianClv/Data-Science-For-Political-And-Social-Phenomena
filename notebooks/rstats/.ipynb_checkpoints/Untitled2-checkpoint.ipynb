# Basic Scatterplot in ggplot2
Original source: ggplot2 book

```r
# load the ggplot2 library
library(ggplot2)

# set the seed so we can reproduce the results
set.seed(1410)
```

```r
# create a variable that is the first 100 rows of the diamonds dataset
dsmall <- diamonds[sample(nrow(diamonds), 100), ]
```

```r
# create a scatterplot of carat vs. price
p <- qplot(carat, price, data = diamonds)
```

```r
# view the scatterplot
p
```

```r
# create a scatterplot with the color determined by the diamonds color
p2 <- qplot(carat, price, data = dsmall, colour = color); p2
```

```r
# create a scatterplot withe the point shape determiend by the cut
p3 <- qplot(carat, price, data = dsmall, shape = cut); p3
```

```r
# create a scatterplot with the point obacity reduced
p4 <- qplot(carat, price, data = diamonds, alpha = I(1/200)); p4
```

```r
# create a scatterplot with a smootling line
p5 <- qplot(carat, price, data = dsmall, geom = c("point", "smooth")); p5
```

```r
# create a scatterplot with a smoothing line but no standard error cone
p6 <- qplot(carat, price, data = dsmall, se = FALSE, geom = c("point", "smooth")); p6
```
